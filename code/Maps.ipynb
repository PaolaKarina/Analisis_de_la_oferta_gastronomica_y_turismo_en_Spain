**VEGETARIAN OPTION - MADRID**

```python
df_gomera = df[(df.province == 'La Gomera')]
```

```python
for i,row in df_gomera.iterrows():
    restaurants = {'location': 
                [row['latitude'], row['longitude']],
               'tooltip': row['vegan_options']}
    
    if row["vegan_options"] == "Y":
        icono = Icon (color = "green",
                     prefix = "fa",
                     icon = "cutlery",
                     icon_color = "white")
    else: 
        icono = Icon (color = "red",
                     prefix = "fa",
                     icon = "cutlery",
                     icon_color = "white")

    marker_vegetarian = Marker(**restaurants, icon = icono)
    marker_vegetarian.add_to(mapa_gomera)
                      
mapa_gomera                      
```
